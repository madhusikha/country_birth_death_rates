```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('country_table.csv')
```

```python
# (i) heatmaps
# (ii) clustermap--groups rows and columns which are similar (as per the values)
```

```python
df
```

```python
# for plotting heatmap, we need a labelled index, in addition to column names
# let us make 'Countries' as index of the dataframe

df = df.set_index('Countries')
```

```python
df.head()
```

```python
sns.heatmap(data=df)

# 'Life expentancy' column has different units than the remaining 4 columns
# so better drop 'Life expentancy' column; 
# other columns are rates, so heatmap would be meaningful with good color shades
```

```python
# let us drop 'Life expentancy' column from the dataframe and plot heatmap

df=df.drop('Life expectancy',axis=1)
```

```python
df.head()
```

```python
sns.heatmap(data=df)
```

```python
plt.figure(dpi=200)
sns.heatmap(data=df,linewidth=0.5,annot=True)  

# for linewidth-- try diff numbeers
# for numbers in each cell, annot=True
```

```python
plt.figure(dpi=200)
sns.heatmap(data=df,linewidth=0.5,annot=True,cmap='viridis') 

# colormap- cmap
```

```python
plt.figure(dpi=200)
sns.clustermap(data=df,linewidth=0.5,annot=True,cmap='viridis') 
```

```python
plt.figure(dpi=200)
sns.clustermap(data=df,linewidth=0.5,annot=True,cmap='viridis',
              col_cluster=False) 
```

```python
sns.clustermap(data=df.corr(),linewidth=0.5,annot=True,cmap='viridis',
              col_cluster=True) 
```
